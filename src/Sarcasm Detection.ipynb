Importing Libraries

```python
import numpy as np
import pandas as pd
```

Reading Data

```python
TRAIN_FILE = "../data/train_data.txt"
data_raw = pd.read_csv(TRAIN_FILE,sep="\t", names=['Index','Label','Tweet'])
data_raw.head()
# need to Remove the Index 0
```

# Data Preprocessing

```python
data_raw = data_raw.drop(0,axis=0)
data_raw.head()
```

Removing @ Mark From Tweets

```python
atmark_filtered_tweets=data_raw[~data_raw.Tweet.str.contains("@[\w]*")]
atmark_filtered_tweets.transpose()
```

Removing Special Characters like emoji & Punctuation marks - here not removing the # value

```python
#character_filtered_tweets = atmark_filtered_tweets.Tweet.str.replace("[^a-zA-Z]", "")
#character_filtered_tweets = atmark_filtered_tweets[atmark_filtered_tweets.Tweet.str.replace("[^a-zA-Z]", "")]
#character_filtered_tweets = atmark_filtered_tweets['Tweet'].str.replace("[^a-zA-Z]", " ")
#character_filtered_tweets = atmark_filtered_tweets['Tweet'.replace("[^a-zA-Z]", " ")]
character_filtered_tweets = atmark_filtered_tweets.replace(to_replace='[^a-zA-Z0-9#]', value='', regex=True)
#filtered_tweets = atmark_filtered_tweets.replace({'Tweets': '^a-zA-Z#'}, {'Tweets': ''}, regex=True)
character_filtered_tweets.transpose()
```

Removing Index Column - Not necessary for predection

```python
data_cleaned = character_filtered_tweets.drop('Index',axis=1)
data_cleaned.transpose()
```

# Defining Independent(X) & Dependent(Y) Variable

```python
X = data_cleaned['Tweet']
Y = data_cleaned['Label']
```

Counter Vectorizing

```python
from sklearn.feature_extraction.text import CountVectorizer
vc = CountVectorizer()
vc.fit(X)
X_coded = vc.transform(X)
X_coded = X_coded.toarray()
print(X_coded)
```

Label Encoder 

```python
from sklearn import preprocessing
label_encode = preprocessing.LabelEncoder()
labels = np.unique(Y)
label_encode.fit(labels)
Y_coded = list(label_encode.transform(Y))
Y_coded = np.asarray(Y_coded)
print(Y_coded)
```

# Model Building - RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_coded, Y_coded, test_size=0.1,random_state=109) # 70% training and 30% test
#print(X_train)
#print(y_test)
model = RandomForestClassifier(n_estimators=3,min_samples_leaf=1)
# Train the model using the training sets
model.fit(X_train,y_train)
# Predicting for Testing Sets
y_pred = model.predict(X_test)
```

# Model Evaluation - RandomForest Classifier

```python
from sklearn import metrics
print("Random Forest Classifier Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

# Model Building - SVM

```python
from sklearn import svm
model_svm = svm.SVC(gamma='scale')
# Train the model using the training sets
model_svm.fit(X_train,y_train)
# Predicting for Testing Sets
y_pred_svm = model.predict(X_test)
```

# Model Evaluation - SVM

```python
print(" SVM Accuracy:",metrics.accuracy_score(y_test, y_pred_svm))
```
